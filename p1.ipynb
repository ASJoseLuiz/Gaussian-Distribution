```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

mu, sigma = 175, 6
data = np.random.normal(mu, sigma, 10000)
data
```

```python
sns.histplot(data, bins=30, kde=True, stat='probability')
plt.xlabel('Heigth (men)')
plt.ylabel('Probability')
plt.title('Distribution of Mens Heigth')
plt.xticks(range(155,200,5))
plt.show()
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Cálculo da média
media = df.mean()
media
```

```python
# Cálculo do desvio padrão
std = df.std()
std
```

```python
# Cálculo da Variância
var = df.var()
var
```

```python
# Coeficiente de variação das alturas
coef = std/media
coef
```

```python
# Assimetria
ass = df.skew()
ass
```

```python
# Curtose
kurt = df.kurtosis()
kurt
```

### Z-SCORE


```python
# Quantos % dos homens tem menos de 183cm?
# Step 1 : Encontrar z-score
# Step 2 : Encontrar a probabilidade calculando a área sob a curva da função densidade de probabilidade

from scipy.stats import norm
zScore = (183 - media)/std
# norm.cdf(zScore) Função Densidade de Probabilidade
p = norm.cdf(zScore)
p
# A porcentagem é algo em torno de 90%
```

```python
# Qual a probabilidade de um homem ser maior que 177cm?
zScore = (177 - media)/std
p = 1 - norm.cdf(zScore)
p
# A probabilidade de um homem ser maior que 177cm é de 36%
```

```python
# Qual é a altura de um homem que é maior que 80% de todos os homens
z = norm.ppf(0.8) # Passagem da porcentagem, retorna a variável aleatória normal Z
x = (z * std) + media # Utilizando a fórmula de zScore encontramos x
x.round()
# A altura arredondada de um homem que é maior que 80% dos homens é 180cm
```
